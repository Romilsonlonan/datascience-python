```python
import matplotlib as mpl 
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *
```

```python
x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

# Definindo os eixos
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_xlabel('y')
axes.set_title('Gráfico de linha');
```

```python
# Grafico com duas figuras
x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

# Eixo da figura principal
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Eixo da figura secundária
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Figura principal 
axes.set_xlabel('x')
axes.set_xlabel('y')
axes.set_title('Gráfico de linha');
```
