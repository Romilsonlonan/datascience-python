```python
# PLOTTING

import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
# Time Series plot
ts = pd.Series(np.random.randn(500), index=pd.date_range('1/1/2015', periods=500))
ts = ts.cumsum()
ts.plot()
```

```python
# DataFrame Plot
df = pd.DataFrame(np.random.randn(500, 6), index = ts.index, columns = ['A', 'B', 'C', 'D', 'E', 'F'])
df = df.cumsum()
plt.figure(); df.plot(); plt.legend(loc = 'best')
```
